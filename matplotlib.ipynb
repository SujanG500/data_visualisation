```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3]
y = [1,2,3]
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x, y, label= "y=mx", color = "red", linestyle= "--", linewidth = 2)
plt.plot(x2, x2**2)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks([1,2,3])
plt.yticks([1,2,3])

plt.legend()
plt.show()
```

```python
def root(x):
    i = 0
    lss = []
    for i in range(0, len(x)):
        y = x[i]**2 + 3
        lss.append(y)
    return lss


y = []
x = [1, 1.5, 2, 2.6]
y = root(x)
print(y)

plt.plot(x, y, label = "square root", linewidth = 1, color = 'green', marker = '.')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("curve of sqaure root")
plt.legend()
plt.show()
        
        
        
```

```python
labels = ['google', 'meta', 'microsoft', 'amazon', 'apple']
users = [50, 10, 20, 30, 6]
bars =plt.bar(labels, users)
plt.figure(figsize= (5,4), dpi = 50)
plt.xlabel("Companies")
plt.ylabel("Users")

patterns = ['/', '*', '//', 'o', '.']
for i in bars:
    i.set_hatch(patterns.pop())    
plt.show()
```
